$\hat{y} = \beta_0 + \beta_1 \cdot x_1 + \beta_2 \cdot x_2 + \ldots + \beta_k \cdot x_k$

$loss = \sum (y_i - \hat{y}_i)^2$

*MSE* - Mean Square Error

*RMSE* - Root Mean Square Error

*MAE* - Mean Absolute Error

```python
e = ys - yp # Error

SSE = (e ** 2).sum() # Sum Square Error
MSE = (e ** 2).mean() # Mean Square Error
RMSE = (e ** 2).mean() ** 0.5 # Root Mean Square Error
SAE = e.abs().sum() # Sum Absolute Error
MAE = e.abs().mean() # Mean Absolute Error

pandas.DataFrame({
    "N": [len(e)],
    "SSE": [SSE],
    "MSE": [MSE],
    "RMSE": [RMSE],
    "SAE": [SAE],
    "MAE": [MAE],
}).T
```

```python
# Funciones de pérdida

def loss_mse(y, yp):
    e = y - yp
    mse = (e ** 2).mean()
    return mse

def loss_mae(y, yp):
    e = y - yp
    mae = e.abs().mean()
    return mae
```

```python
# Optimizador (usando bayesiana)

def optimizer(X, y, beta0, betas, loss):
    beta0_n = beta0 + numpy.random.normal(0, 0.001, 1)
    betas_n = betas.copy() + numpy.random.normal(0, 0.001, len(betas))

    yp1 = beta0 + X.dot(betas)
    yp2 = beta0_n + X.dot(betas_n)

    e1 = loss(y, yp1)
    e2 = loss(y, yp2)

    if e1 > e2:
        return beta0_n, betas_n
    else:
        return beta0, betas
```

```python
beta0 = numpy.random.normal(0, 0.1, 1)
betas = numpy.random.normal(0, 0.1, 15)

for i in range(100_000):
    beta0, betas = optimizer(Xs, ys, beta0, betas, loss_mse)

    yp = beta0 + Xs.dot(betas)

    e = loss_mse(ys, yp)

    if i % 1000 == 0:
        print(i)
        print(beta0, betas)
        print("MSE:", e)
        print("-" * 40)
```

## Modelos de Machine Learning

### Regresión

- Lineal
- Rigdge, Lasso, Elastic-Net
- Soporte Vectorial
- Árbol de Decisión (Regresión)
- Bosque Aleatorio (Regresión)

> Métricas:

* MSE - Error cuadrático medio
* MAE - Error absoluto medio

> Visualización:

* Predicción vs Respuesta (`regplot`) - Simétrica a la recta identidad de 45°
* Error (`densityplot / kdeplot`) - Indica normalidad en el error

### Clasificación

- Curva Logística
- Árbol de Decisión (Clasificación)
- Bosque Aleatorio (Clasificación)
- Boosting por Gradiente Extremo (Clasificación)

> Métricas:

* Matriz de confusión - (Verdero-Negativo [TN], Falso-Positivo [FP], Falso-Negativo [FN], Verdadero-Positivo [TP])
* Exactitud - (TP + TN) / (TN + FP + FN + TP)
* Precisión - TP / (TP + FP)
* Sensibilidad - TP / (TP + FN)
* Especificidad - TN / (TN + FN)
* F1-Score - 2 * Precisión * Sensibilidad / (Precisión + Sensibilidad)

> Visualización:

* Matriz de Confusión (Mapa de calor o `heatmap`)
* AUC / ROC

### Redes Neuronales

- Regresión: Activación Lineal / ReLU
- Clasificación: Sigmoid (Monocategoría) / Softmax (Multicategoría)
