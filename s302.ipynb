> Calcular el area de un triangulo

```python
a = float(input("Lado 1: "))
b = float(input("Lado 2: "))
c = float(input("Lado 3: "))

print(f"A={a} | B={b} | C={c}")

if a < b + c: # <condicional condicion>:
    # BLOQUE1 +++
    if b < a + c: # <condicional condicion>:
        # BLOQUE1.1 +++
        if c < a + b: # <condicional condicion>:
            # BLOQUE1.1.1 +++
            # <sentencia> - asignacion de variable
            s = (a + b + c) / 2
            # <sentencia> - asignacion de variable
            area = (s * (s - a) * (s - b) * (s - c)) ** 0.5
            print(f"AREA: {area:.4f}")
            # BLOQUE1.1.1 ---
        else: # <contra-condicional>
            # BLOQUE1.1.2 +++
            print("El lado C es mas grande o igual que la suma de los otros")
            # BLOQUE1.1.2 ---
        # BLOQUE1.1 ---
    else: # <contra-condicional>
        # BLOQUE1.2 +++
        print("El lado B es mas grande o igual que la suma de los otros")
        # BLOQUE1.2 ---
    # BLOQUE1 ---
else: # <contra-condicional>
    # BLOQUE2 +++
    print("El lado A es mas grande o igual que la suma de los otros")
    # BLOQUE2 ---
```

> Promedio de calificaciones

```python
# Lista de calificaciones vacia
calificaciones = []

# Capturamos la cantidad de calificaciones
N = int(input("Numero de calificaciones (a capturar): "))

# Inicializamos un contador/iterador en 1
i = 1
while len(calificaciones) < N:
    # 1. Capturar la i-esima calificacion
    calificacion = float(input(f"Calificacion {i}: "))
    
    # 2. Retenemos la calificacion en la lista
    # Agregamos la calificacion al final de la lista
    # (inicialmente vacia)
    calificaciones.append(calificacion)

    # 3. Incremento por reemplazo del contador/iterador
    # <nuevo> = <anterior> + <cantidad>
    i = i + 1

print(calificaciones)

print(sum(calificaciones) / len(calificaciones))
```

```python
calificaciones = []

N = int(input("Numero de calificaciones a capturar: "))

# <secuenciador elemento en secuencia>
# range(a, b, s=1) -> [a, a + s, a + 2 * s, ..., <b]
for i in range(1, N + 1):
    calificacion = float(input(f"Calificacion {i}: "))
    calificaciones.append(calificacion)

calificaciones, sum(calificaciones) / len(calificaciones)
```

```python
N = int(input("Numero de calificaciones a capturar: "))

def capturarCalificacion(_):
    calificacion = float(input("Calificacion: "))
    return calificacion

calificaciones = list(map(capturarCalificacion, [0] * N))

calificaciones, sum(calificaciones) / len(calificaciones)
```

```python
import numpy
import pandas

# 1. Determinar el numero de calificaciones
N = int(input("Numero de calificaciones a capturar: "))

# 2. Transformar una calificacion 0 a una dada por el usuario
def capturarCalificacion(_):
    calificacion = float(input("Calificacion: "))
    return calificacion

# 3. Transformar N-ceros en capturas por el usuario
calificaciones = pandas.Series(numpy.zeros(N)).apply(capturarCalificacion)

# 4. Reportar la serie y el promedio
calificaciones, calificaciones.mean()
```

```python
# N - Representa el numero de dimensiones (profundidades)
# Arreglo - Datos dispuestos sobre cada dimension
#   Dim 1: Vectores
#   Dim 2: Matrices
#   ...
#   Dim N: Cubos de datos superiores

# N - dimensiones
# (1D) n - observaciones
# (2D) n x m - observaciones
# (3D) n x m x l - observaciones
# (4D) n x m x l x q - observaciones
# ...
# (ND) n1 x n2 x ... x nN - observaciones

import numpy

# Vector (1D) - 10 observaciones todas cero
x = numpy.zeros(10)

x
```
