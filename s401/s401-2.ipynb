```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

options = Options()

options.add_experimental_option("prefs", {
    "download.default_directory": "/Users/dragonnomada/Desktop/GitHub/ia-cic-abril-2026/s401/descargas",
    "download.prompt_for_download": False,
    "download.dictory_upgrade": True,
    "safebrowsing.enabled": True,
})

navegador = webdriver.Chrome(options=options)
```

```python
navegador.get("https://inegi.org.mx")
sleep(1)
```

> Acceder al elemento `<a ...>Sistemas de consulta</a>`
>
> Representa la pestaña en la barra de navegacion de la pagina web
> que nos lleva al sistema de consulta
>
> XPATH: `//*[@id="navbarSupportedContent"]/div[2]/ul/li[6]/a`

```python
linkSistemaConsulta = navegador.find_element(
    by="xpath", 
    value='//*[@id="navbarSupportedContent"]/div[2]/ul/li[6]/a'
)

linkSistemaConsulta.get_attribute("outerHTML")
```

```python
navegador.execute_script(
    "arguments[0].scrollIntoView({block:'center'})", 
    linkSistemaConsulta
)
sleep(1)
```

```python
linkSistemaConsulta.click()
sleep(1)
```
